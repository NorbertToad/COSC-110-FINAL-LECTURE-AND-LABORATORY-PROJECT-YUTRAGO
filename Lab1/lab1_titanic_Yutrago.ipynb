Question 3:

  - What can you deduce on the survival rate of the passengers?

Answer:

  - As you cna see from the bar graph below, from C, the people were rich, although they were second to last for the most people who embarked. While on S, the place where people boarded the most, the poor people greatly outnumbered the rich and middle class while on Q, almost everyone were poor.
  
  - With this, we can deduct that Q was a place of wealth were people were generally rich. S had the most people so we could say that it had the most population while Q had the poorest economy and the people that lived there were mostly peasants.


```python
import matplotlib.pyplot as plt

plt.hist(df[df["Pclass"] == 1]["Embarked"].dropna(), bins=5, alpha=.5, label="Rich")
plt.hist(df[df["Pclass"] == 2]["Embarked"].dropna(), bins=5, alpha=.4, label="Middle")
plt.hist(df[df["Pclass"] == 3]["Embarked"].dropna(), bins=5, alpha=.3, label="Poor")

plt.legend()
plt.xlabel("Embarked")
plt.ylabel("Count")
plt.show()
```

Question 3:

  - What can you deduce on the survival rate of the passengers?

Answer:
  
  - As you can see from the chart below, the people who died the most were from Class 3, the lowest class. The most survivors out of all of them were the ones from the first class and the lowest survival rate were from the second class. This leads me to believe that Class 3 probably had the most inner or lowest rooms in the ship which would've probably hindered their ability to survive.

```python
import matplotlib.pyplot as plt

plt.hist(df[df["Survived"] == 1]["Pclass"].dropna(), bins=5, alpha=.5, label="Survived")
plt.hist(df[df["Survived"] == 0]["Pclass"].dropna(), bins=5, alpha=.5, label="Not Survived")

plt.legend()
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```
